**Q1 Show the no. of confirmed, death,recovered cases in each region?**

```python
data.groupby('Region').sum()
```

```python
data.groupby('Region')['Confirmed'].sum().sort_values(ascending=False).head(20)
```

```python
data.groupby('Region')['Confirmed','Recovered'].sum().sort_values(by='Confirmed',ascending= False).head(10)
```

**Q2 Remove all the records where confirmed cases are less than 10?**

```python
data[~(data.Confirmed<10)]
```

**Q3 In which region maximum number of confirmed casees were recorded?**

```python
data.groupby('Region').sum().sort_values(by='Confirmed',ascending=False).head(2)
```

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending=False).head(2)
```

**Q4 In which region minimum no. of death were recorded?**

```python
data.groupby('Region').sum().sort_values(by='Deaths',ascending=True).head(2)
```

```python
data.groupby('Region').Deaths.sum().sort_values(ascending=True).head(2)
```

**Q5 How many confirmed, deaths, recovered cases reported in india?**

```python
data[data.Region=='India']
```

**Q6 Sort the entire dataset wrt no. of confirmed cases in ascending order?**

```python
data.sort_values(by='Confirmed',ascending=True).head(50)
```
